```python
#Import Libraries
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
import zipfile
```

```python
local_ref = '573774_1043123_compressed_Musical_Instruments_5.json.zip'
zip_ref = zipfile.ZipFile(local_ref, 'r')
data_file = zip_ref.extractall()
zip_ref.close()
```

```python
#Import Json file
dataset = pd.read_json('Musical_Instruments_5.json', lines = True)
```

```python
#Store all the review texts in sentences
sentences = dataset['reviewText']
```

```python
sentences
```

```python
#Tokenizer gives the index to different words in the sentences
#If a word which is used for testing , is not present in the tokenizer list then it return OOV
tokenizer = Tokenizer(num_words = 1000, oov_token = '<OOV>')
```

```python
#fit the tokenizer to sentences
tokenizer.fit_on_texts(sentences)
```

```python
#Returns the dictionary containing key value pairs
word_index = tokenizer.word_index
```

```python
word_index
```

```python
#sentences will be transformed to the array of lists as the words will be given the nu,bers according to the word indexes
sequences = tokenizer.texts_to_sequences(sentences)
```

```python
sequences
```

```python
#For training neural networks input data should have uniform size so padding will be donw=e to make the input sequences 
#of same size
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded = pad_sequences(sequences, padding = 'post')
```

```python
X = padded
y = dataset.iloc[:,5].values
y = y.reshape(-1,1)
```

```python
#overall rating is converted to categorical features
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(),[0])], remainder = 'passthrough')
y = ct.fit_transform(y).toarray()
```

```python
#Splitting data into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
#Neural Network Model
#Add the embedding layer in the front the network layers
model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding( input_dim = 1000 , output_dim =64, input_length = 2059))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(10, activation = 'relu'))
model.add(tf.keras.layers.Dense(5, activation = 'softmax'))
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['categorical_accuracy'])
model.summary()
```

```python
#Fit the model to the training set and the validation data set
model.fit(X_train, y_train, validation_data = (X_test, y_test))
```
